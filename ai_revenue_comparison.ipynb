# AI Companies vs Traditional Companies: Revenue Potential Comparison

This notebook visualizes the projected revenue potential of AI-focused companies compared to traditional companies.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data: Revenue potential projections (in billions USD)
years = ['2024', '2025', '2026', '2027', '2028', '2029', '2030']

# AI Companies projected revenue growth
ai_companies = {
    'OpenAI': [3.7, 8, 15, 28, 45, 70, 100],
    'Anthropic': [0.9, 2.5, 6, 12, 22, 38, 60],
    'NVIDIA (AI Segment)': [47, 75, 110, 150, 190, 230, 280],
    'Google AI/DeepMind': [30, 45, 65, 90, 120, 155, 200]
}

# Traditional Companies projected revenue growth
traditional_companies = {
    'ExxonMobil': [413, 420, 425, 430, 435, 440, 445],
    'Walmart': [648, 660, 675, 690, 705, 720, 735],
    'Ford': [176, 180, 183, 186, 189, 192, 195],
    'Coca-Cola': [46, 47, 48, 49, 50, 51, 52]
}

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('AI Companies vs Traditional Companies: Revenue Potential', fontsize=16, fontweight='bold')

# Color schemes
ai_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
trad_colors = ['#DDA0DD', '#F0E68C', '#87CEEB', '#FFA07A']

# Plot 1: AI Companies Revenue Growth
ax1 = axes[0, 0]
for i, (company, revenues) in enumerate(ai_companies.items()):
    ax1.plot(years, revenues, marker='o', linewidth=2, label=company, color=ai_colors[i])
ax1.set_title('AI Companies Revenue Growth', fontweight='bold')
ax1.set_xlabel('Year')
ax1.set_ylabel('Revenue (Billions USD)')
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)

# Plot 2: Traditional Companies Revenue Growth
ax2 = axes[0, 1]
for i, (company, revenues) in enumerate(traditional_companies.items()):
    ax2.plot(years, revenues, marker='s', linewidth=2, label=company, color=trad_colors[i])
ax2.set_title('Traditional Companies Revenue Growth', fontweight='bold')
ax2.set_xlabel('Year')
ax2.set_ylabel('Revenue (Billions USD)')
ax2.legend(loc='upper left')
ax2.grid(True, alpha=0.3)

# Plot 3: Growth Rate Comparison (Bar Chart)
ax3 = axes[1, 0]
ai_growth_rates = []
trad_growth_rates = []
for company, revenues in ai_companies.items():
    growth = ((revenues[-1] - revenues[0]) / revenues[0]) * 100
    ai_growth_rates.append(growth)
for company, revenues in traditional_companies.items():
    growth = ((revenues[-1] - revenues[0]) / revenues[0]) * 100
    trad_growth_rates.append(growth)

x = np.arange(4)
width = 0.35
bars1 = ax3.bar(x - width/2, ai_growth_rates, width, label='AI Companies', color='#FF6B6B')
bars2 = ax3.bar(x + width/2, trad_growth_rates, width, label='Traditional Companies', color='#4ECDC4')
ax3.set_title('Projected Growth Rate (2024-2030)', fontweight='bold')
ax3.set_ylabel('Growth Rate (%)')
ax3.set_xticks(x)
ax3.set_xticklabels(['Company 1', 'Company 2', 'Company 3', 'Company 4'])
ax3.legend()
ax3.grid(True, alpha=0.3, axis='y')

# Plot 4: Total Market Potential (Pie Chart)
ax4 = axes[1, 1]
ai_total_2030 = sum([revenues[-1] for revenues in ai_companies.values()])
trad_total_2030 = sum([revenues[-1] for revenues in traditional_companies.values()])
sizes = [ai_total_2030, trad_total_2030]
labels = [f'AI Companies\n${ai_total_2030}B', f'Traditional Companies\n${trad_total_2030}B']
colors = ['#FF6B6B', '#4ECDC4']
explode = (0.05, 0)
ax4.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax4.set_title('Projected 2030 Revenue Share (Selected Companies)', fontweight='bold')

plt.tight_layout()
plt.savefig('ai_revenue_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n" + "="*60)
print("KEY INSIGHTS:")
print("="*60)
print(f"\nAI Companies Average Growth Rate: {np.mean(ai_growth_rates):.1f}%")
print(f"Traditional Companies Average Growth Rate: {np.mean(trad_growth_rates):.1f}%")
print(f"\nAI companies show ~{np.mean(ai_growth_rates)/np.mean(trad_growth_rates):.0f}x higher growth potential!")
```

```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RgbColor
from pptx.enum.text import PP_ALIGN, MSO_ANCHOR
from pptx.enum.shapes import MSO_SHAPE
import matplotlib.pyplot as plt
import numpy as np
from io import BytesIO

# Create presentation
prs = Presentation()
prs.slide_width = Inches(13.333)
prs.slide_height = Inches(7.5)

# Color definitions
AI_COLOR = RgbColor(255, 107, 107)      # #FF6B6B
TRAD_COLOR = RgbColor(78, 205, 196)     # #4ECDC4
ACCENT_COLOR = RgbColor(255, 217, 61)   # #FFD93D
DARK_BG = RgbColor(26, 26, 46)          # #1a1a2e
WHITE = RgbColor(255, 255, 255)

def add_dark_background(slide):
    """Add dark background to slide"""
    background = slide.shapes.add_shape(MSO_SHAPE.RECTANGLE, 0, 0, prs.slide_width, prs.slide_height)
    background.fill.solid()
    background.fill.fore_color.rgb = DARK_BG
    background.line.fill.background()
    # Send to back
    spTree = slide.shapes._spTree
    sp = background._element
    spTree.remove(sp)
    spTree.insert(2, sp)

def add_text_box(slide, left, top, width, height, text, font_size=18, bold=False, color=WHITE, align=PP_ALIGN.CENTER):
    """Add a text box with styling"""
    txBox = slide.shapes.add_textbox(left, top, width, height)
    tf = txBox.text_frame
    tf.word_wrap = True
    p = tf.paragraphs[0]
    p.text = text
    p.font.size = Pt(font_size)
    p.font.bold = bold
    p.font.color.rgb = color
    p.alignment = align
    return txBox

# ============== SLIDE 1: Title Slide ==============
slide1 = prs.slides.add_slide(prs.slide_layouts[6])  # Blank layout
add_dark_background(slide1)

add_text_box(slide1, Inches(0.5), Inches(2), Inches(12.3), Inches(1.5),
             "AI vs TRADITIONAL COMPANIES", font_size=54, bold=True, color=WHITE)
add_text_box(slide1, Inches(0.5), Inches(3.5), Inches(12.3), Inches(1),
             "Revenue Potential Comparison 2024-2030", font_size=32, color=RgbColor(136, 136, 136))
add_text_box(slide1, Inches(0.5), Inches(5.5), Inches(12.3), Inches(0.5),
             "Projected Growth Analysis", font_size=20, color=ACCENT_COLOR)

# ============== SLIDE 2: Key Stats ==============
slide2 = prs.slides.add_slide(prs.slide_layouts[6])
add_dark_background(slide2)

add_text_box(slide2, Inches(0.5), Inches(0.5), Inches(12.3), Inches(1),
             "KEY STATISTICS", font_size=40, bold=True, color=WHITE)

# Stats boxes
stats = [
    ("2,600%", "AI Growth Rate", AI_COLOR),
    ("8%", "Traditional Growth", TRAD_COLOR),
    ("$640B", "AI 2030 Revenue", ACCENT_COLOR),
    ("325x", "Growth Multiple", RgbColor(224, 86, 253))
]

for i, (value, label, color) in enumerate(stats):
    left = Inches(0.8 + i * 3.1)
    # Box shape
    box = slide2.shapes.add_shape(MSO_SHAPE.ROUNDED_RECTANGLE, left, Inches(2.5), Inches(2.8), Inches(3))
    box.fill.solid()
    box.fill.fore_color.rgb = DARK_BG
    box.line.color.rgb = color
    box.line.width = Pt(3)
    # Value
    add_text_box(slide2, left, Inches(3), Inches(2.8), Inches(1.2),
                 value, font_size=48, bold=True, color=color)
    # Label
    add_text_box(slide2, left, Inches(4.3), Inches(2.8), Inches(0.8),
                 label, font_size=18, color=WHITE)

# ============== SLIDE 3: AI Companies ==============
slide3 = prs.slides.add_slide(prs.slide_layouts[6])
add_dark_background(slide3)

add_text_box(slide3, Inches(0.5), Inches(0.3), Inches(12.3), Inches(0.8),
             "AI COMPANIES - Projected 2030 Revenue", font_size=36, bold=True, color=AI_COLOR)

ai_companies_data = [
    ("OpenAI", "$100B", "2,603%"),
    ("Anthropic", "$60B", "6,567%"),
    ("NVIDIA AI", "$280B", "496%"),
    ("Google AI", "$200B", "567%")
]

for i, (name, revenue, growth) in enumerate(ai_companies_data):
    left = Inches(0.8 + i * 3.1)
    # Card
    card = slide3.shapes.add_shape(MSO_SHAPE.ROUNDED_RECTANGLE, left, Inches(1.5), Inches(2.8), Inches(4.5))
    card.fill.solid()
    card.fill.fore_color.rgb = RgbColor(40, 40, 70)
    card.line.color.rgb = AI_COLOR
    card.line.width = Pt(2)
    # Company name
    add_text_box(slide3, left, Inches(1.8), Inches(2.8), Inches(0.6),
                 name, font_size=24, bold=True, color=WHITE)
    # Revenue
    add_text_box(slide3, left, Inches(2.8), Inches(2.8), Inches(1),
                 revenue, font_size=42, bold=True, color=AI_COLOR)
    # Growth label
    add_text_box(slide3, left, Inches(4), Inches(2.8), Inches(0.5),
                 "Growth:", font_size=14, color=RgbColor(136, 136, 136))
    # Growth value
    add_text_box(slide3, left, Inches(4.5), Inches(2.8), Inches(0.8),
                 growth, font_size=28, bold=True, color=ACCENT_COLOR)

# ============== SLIDE 4: Traditional Companies ==============
slide4 = prs.slides.add_slide(prs.slide_layouts[6])
add_dark_background(slide4)

add_text_box(slide4, Inches(0.5), Inches(0.3), Inches(12.3), Inches(0.8),
             "TRADITIONAL COMPANIES - Projected 2030 Revenue", font_size=36, bold=True, color=TRAD_COLOR)

trad_companies_data = [
    ("ExxonMobil", "$445B", "7.7%"),
    ("Walmart", "$735B", "13.4%"),
    ("Ford", "$195B", "10.8%"),
    ("Coca-Cola", "$52B", "13.0%")
]

for i, (name, revenue, growth) in enumerate(trad_companies_data):
    left = Inches(0.8 + i * 3.1)
    # Card
    card = slide4.shapes.add_shape(MSO_SHAPE.ROUNDED_RECTANGLE, left, Inches(1.5), Inches(2.8), Inches(4.5))
    card.fill.solid()
    card.fill.fore_color.rgb = RgbColor(40, 40, 70)
    card.line.color.rgb = TRAD_COLOR
    card.line.width = Pt(2)
    # Company name
    add_text_box(slide4, left, Inches(1.8), Inches(2.8), Inches(0.6),
                 name, font_size=24, bold=True, color=WHITE)
    # Revenue
    add_text_box(slide4, left, Inches(2.8), Inches(2.8), Inches(1),
                 revenue, font_size=42, bold=True, color=TRAD_COLOR)
    # Growth label
    add_text_box(slide4, left, Inches(4), Inches(2.8), Inches(0.5),
                 "Growth:", font_size=14, color=RgbColor(136, 136, 136))
    # Growth value
    add_text_box(slide4, left, Inches(4.5), Inches(2.8), Inches(0.8),
                 growth, font_size=28, bold=True, color=ACCENT_COLOR)

# ============== SLIDE 5: Growth Comparison Chart ==============
slide5 = prs.slides.add_slide(prs.slide_layouts[6])
add_dark_background(slide5)

add_text_box(slide5, Inches(0.5), Inches(0.3), Inches(12.3), Inches(0.8),
             "GROWTH RATE COMPARISON", font_size=36, bold=True, color=WHITE)

# Create chart image
fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor('#1a1a2e')
ax.set_facecolor('#1a1a2e')

companies = ['OpenAI', 'Anthropic', 'NVIDIA AI', 'Google AI', 'ExxonMobil', 'Walmart', 'Ford', 'Coca-Cola']
growth_rates = [2603, 6567, 496, 567, 7.7, 13.4, 10.8, 13.0]
colors = ['#FF6B6B']*4 + ['#4ECDC4']*4

bars = ax.barh(companies, growth_rates, color=colors)
ax.set_xlabel('Growth Rate (%)', color='white', fontsize=12)
ax.tick_params(colors='white')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

for bar, rate in zip(bars, growth_rates):
    ax.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2, 
            f'{rate:,.0f}%' if rate > 100 else f'{rate:.1f}%',
            va='center', color='white', fontsize=10)

plt.tight_layout()
img_buffer = BytesIO()
plt.savefig(img_buffer, format='png', facecolor='#1a1a2e', dpi=150, bbox_inches='tight')
img_buffer.seek(0)
plt.close()

slide5.shapes.add_picture(img_buffer, Inches(1.5), Inches(1.3), width=Inches(10))

# ============== SLIDE 6: Timeline ==============
slide6 = prs.slides.add_slide(prs.slide_layouts[6])
add_dark_background(slide6)

add_text_box(slide6, Inches(0.5), Inches(0.3), Inches(12.3), Inches(0.8),
             "AI REVENUE GROWTH TIMELINE", font_size=36, bold=True, color=ACCENT_COLOR)

years = ['2024', '2025', '2026', '2027', '2028', '2029', '2030']
revenues = ['$82B', '$131B', '$196B', '$280B', '$377B', '$493B', '$640B']

for i, (year, rev) in enumerate(zip(years, revenues)):
    left = Inches(0.5 + i * 1.8)
    # Year circle
    circle = slide6.shapes.add_shape(MSO_SHAPE.OVAL, left + Inches(0.5), Inches(3.5), Inches(0.6), Inches(0.6))
    circle.fill.solid()
    circle.fill.fore_color.rgb = ACCENT_COLOR
    circle.line.fill.background()
    # Year text
    add_text_box(slide6, left, Inches(4.3), Inches(1.6), Inches(0.5),
                 year, font_size=18, bold=True, color=WHITE)
    # Revenue
    add_text_box(slide6, left, Inches(2.2), Inches(1.6), Inches(0.8),
                 rev, font_size=24, bold=True, color=ACCENT_COLOR)

# Timeline line
line = slide6.shapes.add_shape(MSO_SHAPE.RECTANGLE, Inches(0.8), Inches(3.7), Inches(11.7), Inches(0.1))
line.fill.solid()
line.fill.fore_color.rgb = RgbColor(100, 100, 100)
line.line.fill.background()

# ============== SLIDE 7: Key Insights ==============
slide7 = prs.slides.add_slide(prs.slide_layouts[6])
add_dark_background(slide7)

add_text_box(slide7, Inches(0.5), Inches(0.3), Inches(12.3), Inches(0.8),
             "KEY INSIGHTS", font_size=40, bold=True, color=WHITE)

insights = [
    ("Fastest Grower", "Anthropic", "6,567% projected growth", AI_COLOR),
    ("Highest Revenue", "Walmart", "$735B by 2030", TRAD_COLOR),
    ("AI Market Total", "Combined AI", "$640B by 2030", ACCENT_COLOR),
    ("Growth Advantage", "AI vs Traditional", "325x faster growth", RgbColor(224, 86, 253))
]

for i, (title, company, detail, color) in enumerate(insights):
    left = Inches(0.8 + i * 3.1)
    # Card
    card = slide7.shapes.add_shape(MSO_SHAPE.ROUNDED_RECTANGLE, left, Inches(1.5), Inches(2.8), Inches(4.5))
    card.fill.solid()
    card.fill.fore_color.rgb = RgbColor(40, 40, 70)
    card.line.color.rgb = color
    card.line.width = Pt(3)
    # Title
    add_text_box(slide7, left, Inches(2), Inches(2.8), Inches(0.6),
                 title, font_size=18, color=color)
    # Company
    add_text_box(slide7, left, Inches(2.8), Inches(2.8), Inches(0.8),
                 company, font_size=28, bold=True, color=WHITE)
    # Detail
    add_text_box(slide7, left, Inches(4), Inches(2.8), Inches(1),
                 detail, font_size=20, bold=True, color=color)

# Save presentation
pptx_path = 'ai_revenue_comparison.pptx'
prs.save(pptx_path)
print(f"âœ… PowerPoint presentation saved as '{pptx_path}'")
```
